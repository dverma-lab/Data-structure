# 2) Explain the difference between mutable and immutable data types with examples.

mutable and immutable.png

Mutable **example**

```python
my_list = [1, 2, 3]
print(id(my_list))  # Memory address before modification

my_list.append(4)  # Modifying the list
print(my_list)

print(id(my_list))  # Memory address remains the same (modified in-place)
```

Immutable **example**

```python
my_str = "Hello"
print(id(my_str))  # Memory address before modification

my_str = my_str + " World"  # Modification creates a new string
print(my_str)

print(id(my_str))  # Memory address changes (new object created)
```

# 5) Why might you use a set instead of a lists in python ?

A set is an unordered collection of unique elements in Python, while a list is an ordered collection that allows duplicates. Here are some key reasons to use a set instead of a list:

1️⃣ Uniqueness (No Duplicates)
Sets automatically remove duplicates, whereas lists allow repeated values.
Useful when you need a collection of distinct elements.


2️⃣ Faster Lookup & Membership Testing (O(1) vs O(n))
Checking if an element exists in a set is O(1) (constant time) due to hashing.
In contrast, searching in a list takes O(n) (linear time).

3️⃣ Set Operations (Union, Intersection, Difference)
Sets provide built-in operations for set theory, which are not as efficient with lists.

4️⃣ Less Memory Usage
Sets use hash tables, which are memory-efficient compared to lists.
Useful when working with large datasets that require unique values.

✅ Use a Set when:

You need unique elements.
Fast lookups and membership checks are required.
You need to perform set operations efficiently.

✅ Use a List when:

You need ordered elements.
Duplicates are required.
Index-based access is needed.

# 8) What is a hash table, and how does it relate to dictionaries in python ?

A hash table is a data structure that stores key-value pairs and enables fast retrieval, insertion, and deletion of data. It uses a hashing function to map keys to unique memory locations, allowing for quick lookups.



Hashing the Key

A hash function converts a key into a numerical index (hash value).
Example: hash("name") → 123456 (varies per execution).
Storing the Value

The value is stored at the computed index in the hash table.
Handling Collisions

If multiple keys generate the same hash (collision), Python resolves it using open addressing or chaining.
Fast Lookups

Since indexing is direct, retrieving a value is O(1) on average.

Python’s dictionary (dict) is implemented as a hash table. Each key-value pair is stored based on the hashed key, ensuring fast lookups.

Example of a Dictionary (Using a Hash Table)

```python
my_dict = {"name": "Deepak", "age": 25, "city": "New York"}
print(my_dict["name"])  # O(1) lookup
```

# 10) Explain why string are immutable in python.

A string (str) in Python is immutable, meaning once it is created, it cannot be changed. Any operation that appears to modify a string actually creates a new string object.

Reasons for String Immutability

1️⃣ Memory Efficiency (String Interning)
Python interns (caches) short and commonly used strings to save memory.
If strings were mutable, changing one reference would unexpectedly modify others.

2️⃣ Hashability (Allows Strings as Dictionary Keys)
Strings are hashable, meaning they can be used as keys in dictionaries.
If strings were mutable, their hash values would change, breaking dictionary lookups.

3️⃣ Thread-Safety (No Unexpected Modifications)
Multiple threads can safely share string objects without worrying about unintended changes.
This makes Python more efficient for concurrent programming.

4️⃣ Simplicity and Predictability
Making strings immutable prevents accidental modifications.
Any modification creates a new object, ensuring data integrity.

# 11) What advantages do dictionaries offer over lists for certain tasks ?

Dictionaries (dict) and lists (list) are both commonly used data structures, but dictionaries offer several advantages over lists for certain tasks.

2️⃣ Key-Value Pair Storage (More Meaningful Data Representation)

3️⃣ No Need for Index-Based Searches
Lists: Require knowing the exact index or looping through elements.
Dictionaries: Allow direct access using keys, avoiding loops.

4️⃣ Efficient Data Organization and Grouping
Dictionaries allow grouping related data efficiently.
Lists don’t have this structure and require extra steps.

5️⃣ Flexibility with Heterogeneous Data
Dictionaries can store complex and nested data structures.
Lists store items in sequence, making them less flexible.

If data doesn’t require unique keys, lists are simpler and more straightforward.

# 19) Why are dictionaries considered unordered,and how does that affect data retrieval ?

Before Python 3.7, dictionaries (dict) were considered unordered because they did not maintain the insertion order of key-value pairs. The order of elements could change due to the internal hash table structure used for storing data.

However, starting from Python 3.7+, dictionaries maintain insertion order, meaning items are stored in the order they were added.


 This Affect Data Retrieval :-

1️⃣ Before Python 3.7 (Unordered Dictionaries)
The order of items was not guaranteed.
Retrieving elements would not always follow insertion order.

2️⃣ Python 3.7+ (Ordered Dictionaries)
The dictionary remembers the order of insertion.
When iterating over a dictionary, items appear in the same order they were added.
